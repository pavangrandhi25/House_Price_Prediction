```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/pavangrandhi25/Datasets/main/Houses_Vijayawada.csv')
data
```

```python
data.info()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.info()
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
data['Location'] = encoder.fit_transform(data['Location'])

encoded_to_original = dict(zip(encoder.transform(encoder.classes_),
                               encoder.classes_))

for encoded_label, original_value in encoded_to_original.items():
  print(f'{encoded_label} - {original_value}')
```

```python
data['status'] = encoder.fit_transform(data['status'])

encoded_to_original = dict(zip(encoder.transform(encoder.classes_),
                               encoder.classes_))

for encoded_label, original_value in encoded_to_original.items():
  print(f'{encoded_label} - {original_value}')
```

```python
data['Facing'] = encoder.fit_transform(data['Facing'])

encoded_to_original = dict(zip(encoder.transform(encoder.classes_),
                               encoder.classes_))

for encoded_label, original_value in encoded_to_original.items():
  print(f'{encoded_label} - {original_value}')
```

```python
data['Type'] = encoder.fit_transform(data['Type'])

encoded_to_original = dict(zip(encoder.transform(encoder.classes_),
                               encoder.classes_))

for encoded_label, original_value in encoded_to_original.items():
  print(f'{encoded_label} - {original_value}')
```

```python
x = data.drop('Price',axis=1)
y = data['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.3,
                                                    random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
bed = int(input('Enter No. of bedrooms : '))
bath = int(input('Enter No of bathrooms : '))
loc = int(input('Enter location : '))
size = int(input('Enter Sqft : '))
status = int(input('Enter status : '))
facing = int(input('Enter facing direction : '))
Type = int(input('Enter type of property : '))
```

```python
import numpy as np
input_data = np.array([[bed,bath,loc,size,status,facing,Type]])

predicted_price = model.predict(input_data)[0]
print(f'Predicted Price : ₹ {predicted_price} Lakhs')
```
